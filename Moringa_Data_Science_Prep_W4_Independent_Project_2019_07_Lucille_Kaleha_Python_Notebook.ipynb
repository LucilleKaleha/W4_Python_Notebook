## Week 4 Independent Project: Processing Data for Stations to understand Electric Car Usage Over Time

# 1.0 Importing Libraries

```
import pandas as pd
```

# 1.1 Reading and Previewing our Dataset fro the CSV File

```
url = 'http://bit.ly/Autolibdataset'
Autolib = pd.read_csv(url)
Autolib.head()
```

# 1.2 Getting Information about our Dataset

```
# Knowing more about the dataset by accessing its information
Autolib.info()
```

```
# Checking for null values
Autolib.isnull().any()
```

```
# Checking for missing values
Autolib.count()
rows = Autolib.shape[0]
missing = rows - Autolib.count()
missing
```

#1.3 Cleaning the Dataset

```
# Checking for Outliers
Q1 = Autolib.quantile(0.25)
Q3 = Autolib.quantile(0.75)

IQR = Q3 - Q1 

(Autolib < (Q1 - 1.5 * IQR)) |(Autolib > (Q3 + 1.5 * IQR))
# Will not drop any outliers
```

```
# Deleting Irrelevant Columns
Autolib.drop(['Unnamed: 0', 'Address','Cars', 'Charge Slots', 'ID', 'Geo point','Displayed comment',
             'Scheduled at', 'year', 'month'], axis = 1, inplace = True)
Autolib
```

```
# Removing Special Characters from columns
Autolib['Public name'] = Autolib['Public name'].str.replace("/", " ")
```

```
# Dropping the duplicate values in our dataset
Autolib.drop_duplicates()
Autolib.head()
```

```
# Creating new columns with counter difference
Autolib['Bluecar counter difference'] = Autolib['Bluecar counter'].diff()
Autolib['Utilib counter difference'] = Autolib['Utilib counter'].diff()
Autolib['Utilib 1.4 counter difference'] = Autolib['Utilib 1.4 counter'].diff()
Autolib.head()
# A negative value in the counter difference means a car was picked from the station and a positive value means
# a car was returned
```

```
Autolib.columns = ['Bluecar_counter',	'Utilib_counter',	'Utilib_1.4_counter',	'Charging_status', 'City',	'Kind',
        'Postal_code',	'Public_name',	'Rental_status', 'Slots',	'Station_type',	'Status',	'Subscription_status',
        'day',	'hour',	'minute',	'Bluecar_counter_difference', 'Utilib_counter_difference',	'Utilib1.4_counter_difference']
Autolib.head()
```

# 1.4 Answering Questions

Questions Relating to Bluecar Counter

```
# Research Question
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris 
# over the month of April 2018

Autolib[(Autolib.City == 'Paris') & (Autolib.Bluecar_counter_difference < 0)].groupby('hour')['hour'].count().sort_values(ascending = False).head(10)
```

```
# Bonus Question 1
# What is the most popular hour for returning cars?

Autolib[Autolib.Bluecar_counter_difference > 0].groupby('hour')['hour'].count().sort_values(ascending = False).head(10)
```

```
# Bonus Question 2.1
# What station is the most popular?
# Overall?
Autolib[(Autolib.Kind == 'STATION') & (Autolib.Status == 'ok')].groupby('Public_name')['Public_name'].count().sort_values(ascending= False).head(10)
```

```
# Bonus Question 2.2
# What station is the most popular?
# At the most popular picking hour?
Autolib[(Autolib.Kind == 'STATION') & (Autolib.Status == 'ok') & (Autolib.hour == 4)].groupby('Public_name')['Public_name'].count().sort_values(ascending= False).head(10)
```

```
# Bonus Question 3.1
# What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
# Overall?
Autolib[(Autolib.Bluecar_counter_difference < 0)].groupby('Postal_code')['Postal_code'].count().sort_values(ascending = False).head(10)
```

```
# Bonus Question 3.2
# What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
# At the most popular picking hour?
Autolib[(Autolib.Bluecar_counter_difference < 0) & (Autolib.hour == 4)].groupby('Postal_code')['Postal_code'].count().sort_values(ascending = False).head(10)
```
